```python
import google.generativeai as genai              # To import google_generative
from IPython.display import display, Markdown    # The IPython.display module provides utilities for rich media display
                                                 # The Markdown class allows you to render Markdown content dynamically.
import PIL.Image                                 # The PIL module, which stands for Python Imaging Library
```

```python
from google.colab import userdata
GOOGL_API_KEY = userdata.get('GEMINI_API_KEY')
```

```python
genai.configure(api_key=GOOGL_API_KEY)   # This line is added to set the API key
```

```python
image_path = "/content/AI.jpeg"
image = PIL.Image.open(image_path)
image
```

```python
prompt = """Describe the image in detail """
```

```python
print(dir(genai))       # To get  its attributes and methods
```

```python
model = genai.GenerativeModel('gemini-2.0-flash-exp')

response = model.generate_content([prompt, image])
Markdown(response.text)
```
