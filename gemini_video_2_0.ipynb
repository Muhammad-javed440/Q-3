# Video understanding

```python
import google.generativeai as genai
import cv2
import os
from PIL import Image
import io
from google.colab import files,userdata

# API Key Setup
GOOGLE_API_KEY= userdata.get('GEMINI_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

#  Upload Video
uploaded = files.upload()
video_file_path = list(uploaded.keys())[0]

# Function to extract video frames
def extract_frames(video_path, num_frames=5):
        """Extracts a specified number of frames from a video."""
        frames = []
        cap = cv2.VideoCapture(video_path)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        if total_frames <=0:
             raise ValueError(f"Can't load the video with path: {video_path}")

        if num_frames >= total_frames:
          frame_numbers = list(range(total_frames))
        else :
          frame_numbers = [int(i * total_frames / (num_frames)) for i in range(num_frames)]

        for frame_number in frame_numbers:
            cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
            ret, frame = cap.read()
            if ret:
              # Convert frame to RGB
              frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
              # Convert frame to a PIL image
              image = Image.fromarray(frame_rgb)
              frames.append(image)
        cap.release()
        return frames

# Get frames
num_frames_to_extract = 5
try:
    video_frames = extract_frames(video_file_path, num_frames=num_frames_to_extract)
except ValueError as e:
    print(e)


# Prepare Gemini Prompt
prompt = "Describe the main objects and events that are taking place in this video."


# Initialize the Gemini model
model = genai.GenerativeModel('gemini-2.0-flash-exp')

# Make the Gemini request
if video_frames:
   response = model.generate_content([prompt, *video_frames])
   print(response.text)
else:
   print("No video frames extracted.")
```

# **Search as a tool**

```python
from google import genai
client = genai.Client(
    api_key=GEMINI_API_KEY,
    http_options= {'api_version': 'v1alpha'}
  )
MODEL = "gemini-2.0-flash-exp"
```

```python
import asyncio
import base64
import contextlib
import datetime
import os
import json
import wave
import itertools

from IPython.display import display, Audio

from google import genai
from google.genai import types

async def async_enumerate(it):
  n = 0
  async for item in it:
    yield n, item
    n +=1
```

```python
@contextlib.contextmanager
def wave_file(filename, channels=1, rate=24000, sample_width=2):
    with wave.open(filename, "wb") as wf:
        wf.setnchannels(channels)
        wf.setsampwidth(sample_width)
        wf.setframerate(rate)
        yield wf
```

```python
config={
    "generation_config": {"response_modalities": ["AUDIO"]}}

async with client.aio.live.connect(model=MODEL, config=config) as session:
  file_name = 'db.wav'
  with wave_file(file_name) as wav:
    message = "Hello? Gemini are you know imran khan?"
    print("> ", message, "\n")
    await session.send(message, end_of_turn=True)

    turn = session.receive()
    async for n,response in async_enumerate(turn):
      if response.data is not None:
        wav.writeframes(response.data)

        if n==0:
          print(response.server_content.model_turn.parts[0].inline_data.mime_type)
        print('.', end='')


display(Audio(file_name, autoplay=True))
```
